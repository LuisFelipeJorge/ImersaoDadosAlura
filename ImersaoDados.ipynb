Correlação pode ser um indicativo da influência de uma variável sobre outra, quanto mais próximo de 1 estiver o módulo da correlação, maior será a proximidade do comportamento de uma varável em relação ao valor de outra.

Correlação positiva -> a medida que uma variável cresce, a outra cresce junto.

Correlação negativa -> a media que uma variável cresce, a outra diminui.

### Correlação NÃO é Causa

Pode ser um indicativo, para estimular uma investigação mais aprofundada que verifique realmente a presença de uma influência.

## Matrix de correlação

código fonte: https://seaborn.pydata.org/examples/many_pairwise_correlations.html

```python
import numpy as np

gene_correlation_data = data.loc[:, 'g-0':'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(gene_correlation_data, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio

# obs: retiramos o vmax
sns.heatmap(gene_correlation_data, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Implementação do modelo de regressão logística nos dados para prever o MOA de uma determinada assinatura celular (gene e viabilidade celular).
Utilizaremos a biblioteca Sci kit learn.


```python
merged_data_with_results.select_dtypes('float64')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x=merged_data_with_results.select_dtypes('float64')
y=merged_data_with_results['active_moa']

x_trainning, x_test, y_trainning, y_test = train_test_split(x, y, test_size=0.2,stratify=y , random_state=376)

logistic_regression_model = LogisticRegression(max_iter=1000)
logistic_regression_model.fit(x_trainning, y_trainning)
logistic_regression_model.score(x_test, y_test)
```

```python
from sklearn.dummy import  DummyClassifier
from sklearn.metrics import accuracy_score

dummy_model = DummyClassifier('most_frequent')
dummy_model.fit(x_trainning, y_trainning)
dummy_prediction = dummy_model.predict(x_test)
accuracy_score(y_test, dummy_prediction)
```

Modelo Árvore de Decisão


```python
from sklearn.tree import DecisionTreeClassifier


x=merged_data_with_results.select_dtypes('float64')
y=merged_data_with_results['active_moa']

x_trainning, x_test, y_trainning, y_test = train_test_split(x, y, test_size=0.2,stratify=y , random_state=376)

tree_model = DecisionTreeClassifier(max_depth=3)
tree_model.fit(x_trainning, y_trainning)
tree_model.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier


x=merged_data_with_results.select_dtypes('float64')
y=merged_data_with_results['active_moa']

x_trainning, x_test, y_trainning, y_test = train_test_split(x, y, test_size=0.2,stratify=y , random_state=376)

test=[]
trainning=[]

for i in range(1, 15):
  tree_model = DecisionTreeClassifier(max_depth=i)
  tree_model.fit(x_trainning, y_trainning)
  test.append(tree_model.score(x_test, y_test))
  trainning.append(tree_model.score(x_trainning, y_trainning))
  
  
```

```python
test
```

```python
trainning
```

Temos aqui a visualização do fenômeno *OverFeeding*

Em que nosso modelo, a medida que aumentamos a profundidade, fica cada vez melhor em acertar os casos de treino, mas perde a capacidade de genetalizar os resultados e, portanto, falha nas previsões de amostras inéditas

```python
sns.lineplot(x=range(1,15), y=test, label='test')
sns.lineplot(x=range(1,15), y=trainning, label='trainning')

plt.show()
```

Para obeter um resultado melhor, utilizaremos um conjunto de resultados de árvores de decisão, a **Random Forest**

Para isso precisamos converter os valores string para float, pois a biblioteca trabalha apenas com valore numericos

```python
pd.get_dummies(merged_data_with_results['tratamento'])
```

```python
pd.get_dummies(merged_data_with_results, columns=['tratamento', 'dose', 'tempo'])
```

```python
merged_data_with_results.drop(['id', 'number_of_moa', 'active_moa', 'composto'], axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier

x=merged_data_with_results.drop(['id', 'number_of_moa', 'active_moa', 'composto'], axis=1)
x=pd.get_dummies(x, columns=['tratamento', 'dose', 'tempo'])
y=merged_data_with_results['active_moa']

x_trainning, x_test, y_trainning, y_test = train_test_split(x, y, test_size=0.2,stratify=y , random_state=376)

random_forest_model = RandomForestClassifier()
random_forest_model.fit(x_trainning, y_trainning)
random_forest_model.score(x_test, y_test)
```
